Data seems to be imbalanced. Lets handle this Imbalanced data with SMOTE technique.

Before performing any smote, we need to split the data into train n test then apply smote on top of train data.

```python
data.iloc[35985]
```

```python
x = data.loc[:,data.columns != 'strength']
y= data.strength
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_tra,x_te,y_tra,y_te = train_test_split(x,y,test_size=0.3,shuffle=True,random_state=42)
```

```python
def feature_word_division(word):
    char=[]
    for i in word:
        char.append(i)
    return char
```

```python
feature_word_division('irFj8bzUwNALkw8o')
```

```python
# converting password feature into numerical vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vector = TfidfVectorizer(tokenizer = feature_word_division)
x_train = vector.fit_transform(np.array(x_tra['password']))
x_test = vector.transform(np.array(x_te['password']))
```

# Logistic Regression with OVR

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

log_res =  LogisticRegression(penalty = 'l2' , multi_class = 'ovr')
log_res.fit(x_train_sm,y_train_sm)
```

```python
log_res.score(x_test,y_test)
```
